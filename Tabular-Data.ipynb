# Working with Tabular Data

```python
import os
import numpy
```

```python
distance_file = os.path.join('data', 'distance_data_headers.csv')
```

```python
# use the numpy function genfromtxt to read in the data
help(numpy.genfromtxt)
```

```python
distances = numpy.genfromtxt(fname=distance_file, delimiter = ',', dtype = 'unicode')

# delimiter is what is separation elements of the file (ie: commas, space, tabs)
```

```python
print(distances)
```

```python
headers = distances[0]
print(headers)
```

```python
# Take a slice of the data that is just the numerical values. Call this slice data.
data = distances[1:]
print(data)
```

```python
data = data.astype(numpy.float)
```

```python
print(data)
```

```python
# array_name[row,column]
print(data[2,1])
```

```python
# what will be the output of this code? Write a print statment to check yourself.
element1 = data[0,1]
element2 = data[1,0]
```

```python
print(F'The value of element1 is {element1}.')
```

```python
print(F'The value of element1 is {element2}.')
```

```python
#new_array = array_name[start:end, start:end]
small_data = data[0:10,0:3]
# [:,:] every row and column
print(small_data)
```

```python
#Think about what these slices are. Write a print statement to check yourself.
array1 = small_data[5,:]
array2 = small_data[:,1:]
print(F'The value of array1 is {array1}.')
print(F'The value of array2 is {array2}.')
```

```python
print(data)
```

```python
thr4_atp = data[:,1]
average = numpy.mean(thr4_atp)
print(average)
```

```python
# range for loop: range(start,end)
```

```python
# Using commands you already know, calculate the number of columns
num_columns = len(data[0,1:])
print(num_columns)
```

```python
for num in range(1,num_columns):
    column = data[:,num]
    avg_col = numpy.mean(column)
    print(F'{headers[num]}: {avg_col:3f}')
```
